# Statistiques descriptives sur le nombre de mots par paquet de phrases

```python
df_phrases_mots = df["phrases_paquet"].str.split().str.len()
df_phrases_mots.describe()
```

```python
df_phrases_mots
```

```python
print(df_phrases_mots[df_phrases_mots < 100].count())
print(df_phrases_mots[df_phrases_mots > 1200].count())
print(df_phrases_mots[df_phrases_mots > 1500].count())
```

- Nous pouvons constater apres la vérification de la distribution du nombres de mots par paquet :

-premierement, le corpus est composé de 2671 paquets de phrases.

-la moyenne de mots par paquet est de 1566 mots avec une médiane à 1520 mots, ce qui indique que la distribution est légèrement asymétrique vers les valeurs plus élevées mais homogène.

- pour ce qui est des valeurs extrêmes :

 17 / 2671 ≈ 0,64 % c'est négligeable

 6 / 2671 ≈ 0,22 % c'est encore une fois négligeable 

 Nous pouvons les ignorer pour l'analyse ainsi que le pretraitement du corpus. Cependant, il est important de les garder à l'esprit pour les étapes suivantes, notamment lors de la vectorisation et de la modélisation, afin de s'assurer que ces paquets ne biaisent pas les résultats.
